# ต่อตาราง

- ต่อในแนวแกน Y (เพิ่มจำนวน)
- ต่อในแนวแกน X (เพิ่มรายละเอียดของข้อมูล)

```python
BKK = data_df[data_df['province_of_onset'] == 'กรุงเทพมหานคร'] #ตรวจเจอที่ กทม
BKK.head()
```

```python
KKC = data_df[data_df['province_of_onset'] == 'ขอนแก่น'] #คนที่ตรวจเจอที่ขอนแก่น
KKC.head()
```

## ต่อตารางให้เป็นข้อมูล 2 จังหวัด

```python
dataBKKKKC = pd.concat([BKK,KKC]) #concat แลว่าเอามาต่อกันในแนวแกน Y #เอาตาราง KKC มาต่อกันตาราง BKK
dataBKKKKC
```

## ถ้าเอาตารางไม่เหมือนกันมา concat กัน

```python
BKK2 = data_df[data_df['province_of_onset'] == 'กรุงเทพมหานคร'][['sex','age','nationality']] #ของ กทม 
BKK2.head() #เอาแค่ข้อมูลด้านบน
```

```python
KKC2 = data_df[data_df['province_of_onset'] == 'ขอนแก่น'][['age','nationality','sex']] #ของขอนแก่นแต่สลับ
KKC2.head() #เอาแค่ข้อมูลด้านบน
```

```python
pd.concat([BKK2,KKC2]) #เอาตาราง BKK2 กับ KKC2 มาต่อกัน
```

```python
KKC3 = data_df[data_df['province_of_onset'] == 'ขอนแก่น'][['age','nationality','risk']] #ของขอนแก่นเพิ่ม ความเสี่ยงเข้า
KKC3.head()
```

```python
pd.concat([BKK2,KKC3]) #มันเอามาต่อตารางได้แต่ค่าตรงไหนไม่มีจะเป็นค่า NaN จะไม่เอามาใส่มั่ว
```

```python
KKC4 = data_df[data_df['province_of_onset'] == 'ขอนแก่น'][['age','nationality']] #ของขอนแก่นมีแค่ อายุ กับสัญชาติ
KKC4.head()
```

```python
pd.concat([BKK2,KKC4]) #เอาตาราง BKK2 กับ KKC4 มาต่อกัน #มันเอามาต่อตารางได้แต่ค่าตรงไหนไม่มีจะเป็นค่า NaN จะไม่เอามาใส่มั่ว
```

## ต่อแกน X

```python
personal_data = data_df[['No.','sex','age','nationality']] #ข้อมูลส่วนบุคคล ที่สามารถเอามาเชื่อมกันได้
personal_data
```

```python
covid_data = data_df[['No.','announce_date','province_of_isolation','province_of_onset','risk']].sample(frac=1) #ข้อมูลเบื้องต้น #.sample(frac=1) สลับแถวกัน สุ่ม
covid_data.head()
```

## เอาตาราง 2 อันมาต่อกัน

```python
personalcovid_data = personal_data.merge(covid_data) #ทำการต่อตารางกัน
personalcovid_data.head() #เอามาแค่ด้านบน
```

```python
personalcovid_data = covid_data.merge(personal_data) #ทำการต่อตารางกันลองเอา covid_data ขึ้นก่อน
personalcovid_data.head() #เอามาแค่ด้านบน
```
